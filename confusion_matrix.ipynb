```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 1. Define the values for the matrix (Using a COVID-19 example)
tn = 55  # True Negatives
fp = 3   # False Positives
fn = 5   # False Negatives
tp = 42  # True Positives

# Create the confusion matrix array
cm = np.array([[tn, fp],
               [fn, tp]])

# 2. Create labels with custom text and counts
group_names = ['True Negatives (TN)\n(Correctly Predicted Healthy)',
               'False Positives (FP)\n(Incorrectly Predicted Sick)',
               'False Negatives (FN)\n(Missed Sick Patients)',
               'True Positives (TP)\n(Correctly Predicted Sick)']

group_counts = [f'{value:0.0f}' for value in cm.flatten()]
labels = [f'{v1}\n\nCount: {v2}' for v1, v2 in zip(group_names, group_counts)]
labels = np.asarray(labels).reshape(2, 2)

# 3. Set up the figure size and aesthetics for a video-friendly display
plt.figure(figsize=(10, 7))

# Use a clean, professional color map (Blues for technical/medical context)
ax = sns.heatmap(cm, annot=labels, fmt='', cmap='Blues',
                 annot_kws={'size': 12, 'weight': 'bold'},
                 xticklabels=['Predicted Negative\n(Healthy)', 'Predicted Positive\n(Sick)'],
                 yticklabels=['Actual Negative\n(Healthy)', 'Actual Positive\n(Sick)'],
                 cbar=True, linewidths=2, linecolor='white')

# 4. Add formatting and labels
plt.title('COVID-19 Diagnostic Confusion Matrix', fontsize=16, pad=20, weight='bold')
plt.ylabel('True (Actual) Label', fontsize=12, labelpad=10)
plt.xlabel('Predicted Label', fontsize=12, labelpad=10)

# Rotate tick marks for better readability
plt.xticks(rotation=0)
plt.yticks(rotation=0)

# 5. Display the plot
plt.tight_layout()
plt.show()
```
