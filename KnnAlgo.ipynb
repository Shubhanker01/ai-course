```python
# code for Knn algorithm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# create dataset
np.random.seed(0)
apple_weights = np.abs(np.random.normal(500,40,15))
apple_prices = np.random.normal(4.5,0.6,15)

watermelon_weights = np.abs(np.random.normal(7000,400,15))
watermelon_prices = np.random.normal(10,0.8,15)

lemon_weights = np.abs(np.random.normal(80,10,15))
lemon_prices = np.random.normal(1.5,0.4,15)

# combine dataset
X = np.vstack((
    np.column_stack((apple_weights,apple_prices)),
    np.column_stack((watermelon_weights,watermelon_prices)),
    np.column_stack((lemon_weights,lemon_prices))
))
X_Scaled = scaler.fit_transform(X)
y = np.array(
    ['apple']*15 + ['watermelon']*15 + ['lemon']*15
)

# train data
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_Scaled,y)

# predict a new fruit
new_fruit = np.array([[3000,11.5]])
prediction = knn.predict(scaler.transform(new_fruit))
print(f"prediction: {prediction}")

# accuracy prediction
accuracy = knn.score(X_Scaled,y)
print(f"accuracy: {accuracy}")

# plot
plt.figure(figsize=(8,6))
plt.scatter(apple_weights,apple_prices,color='red',label='apple')
plt.scatter(watermelon_weights,watermelon_prices,color='green',label='watermelon')
plt.scatter(lemon_weights,lemon_prices,color='yellow',label='lemon')
plt.scatter(new_fruit[0][0],new_fruit[0][1],color='blue',label='new fruit')
plt.xlabel('weight')
plt.ylabel('price')
plt.legend()
plt.show()


```
