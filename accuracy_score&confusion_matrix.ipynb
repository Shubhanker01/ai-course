```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
# true labels
x_true = [0,0,1,0,1,1]
# predicted labels
x_predicted = [0,1,0,0,0,1]
acc = accuracy_score(x_true, x_predicted)
print(acc*100)
cf = confusion_matrix(x_true, x_predicted)
print(cf)
```
