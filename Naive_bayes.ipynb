```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
emails = [
    "Win free prizes starting from today",
    "Are you playing tomorrow",
    "One laptop free on buying two",
    "Play and win exciting offers",
    "Can we go for coffee tomorrow?",
    "Buy one get one free sale tomorrow",
    "Hurry bitcoins selling tomorrow",
    "Can you come tomorrow?",
    "Push the code in github asap"
]
labels = ["spam","not spam","spam","spam","not spam","spam","spam","not spam","not spam"]
# convert text to numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)
print(vectorizer.get_feature_names_out())
X.toarray()

# train model
model = MultinomialNB()
model.fit(X,labels)
# test email
test_email = ["Cars are free"]
test_vector = vectorizer.transform(test_email)
predictions = model.predict(test_vector)
print(predictions[0])
```
