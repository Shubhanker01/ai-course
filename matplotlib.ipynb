```python
# using matplotlib
import matplotlib.pyplot as plt
x = [8,9,12]
y = [9,20,11]
plt.plot(x,y)
plt.show()
```

```python
# using bar graph
import matplotlib.pyplot as plt
categories = ['ice cream','bread','dahi']
persons = [12,8,7]
plt.title('Person and dishes They like')
plt.bar(categories,persons)
plt.show()
```

```python
x = [12.9,5.8,9.3]
y = [19.1,8.7,3.33]
plt.scatter(x,y)
plt.show()
```

```python
# drawing pie chart
labels = ['Pune','Chandigarh','Mumbai','Kolkata']
people = [10000,90000,87000,72383]
plt.pie(people,labels=labels,autopct='%1.1f%%')
plt.show()
```
