```python
from catboost import CatBoostClassifier
import pandas as pd
import matplotlib.pyplot as plt
# Sample Data: [Age, City, Gender]
train_data = [
    [25, "New York", "F"],
    [30, "London", "M"],
    [45, "Mumbai", "F"],
    [22, "London", "F"],
    [33, "New York", "M"]
]
# Labels: 1 = Bought, 0 = Didn't Buy
train_labels = [1, 0, 1, 0, 1]
cat_features = [1,2]
model = CatBoostClassifier(iterations=10,learning_rate=0.1,depth=2)
model.fit(train_data,train_labels,cat_features=cat_features,verbose=False)
test_data = [[23,"New York","M"]]
pred = model.predict(test_data)
print("Prediction",pred)

```
