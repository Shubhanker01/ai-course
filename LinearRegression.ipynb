```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

# dataset for banglore house pricing table
data = [
    {'Area_sqfeet':900,'price':550000},
    {'Area_sqfeet':1200,'price':760500},
    {'Area_sqfeet':1500,'price':1110900},
    {'Area_sqfeet':1800,'price':1250000},
    {'Area_sqfeet':2000,'price':1580000},
    {'Area_sqfeet':2200,'price':2050000},
    {'Area_sqfeet':2400,'price':2104000},
    {'Area_sqfeet':2600,'price':2250000},
    {'Area_sqfeet':2800,'price':2500000},
    {'Area_sqfeet':3000,'price':3050000},
    {'Area_sqfeet':3200,'price':3100500},
    {'Area_sqfeet':3400,'price':3250000},
    {'Area_sqfeet':3600,'price':3400000},
    {'Area_sqfeet':3800,'price':3850000},
    {'Area_sqfeet':4000,'price':4900000},
]

df = pd.DataFrame(data)
print(df)

sizes = df['Area_sqfeet']
prices = df['price']
print(sizes)
print(prices)
```

```python
# plot the data
plt.figure(figsize=(10, 6))
plt.scatter(sizes,prices,color='blue',marker='o',label='Data Points')
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price')
plt.title('House Size vs Price')
plt.legend()
plt.grid(True)
# plt.show()
```

```python
# model training part
model = LinearRegression()
model.fit(sizes.values.reshape(-1,1),prices)
predictions = model.predict(sizes.values.reshape(-1,1))
print(predictions)
# mean square error
mse = mean_squared_error(prices,predictions)
print(f"Mean Squared Error: {mse}")
# plot predictions
plt.scatter(sizes,prices,color='blue',marker='o',label='Data Points')
plt.plot(sizes,predictions,color='red',linewidth=2,label='Linear Regression Line')
```
