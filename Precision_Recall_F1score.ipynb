```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt

# ---- Step 1: Create synthetic data (100 people) ----
np.random.seed(42)

actual = np.random.randint(0, 2, 100)      # actual cancer status (ground truth)
predicted = np.random.randint(0, 2, 100)   # model predicted cancer status

# ---- Step 2: Calculate metrics using sklearn ----
precision = precision_score(actual, predicted)
recall = recall_score(actual, predicted)
f1 = f1_score(actual, predicted)

# ---- Step 3: Print results ----
print("Actual:   ", actual)
print("Predicted:", predicted)

cm = confusion_matrix(actual, predicted)

# ---- Step 3: Plot confusion matrix ----
plt.figure(figsize=(6, 6))
plt.imshow(cm)  # matrix visualization

plt.title("Confusion Matrix")
plt.xlabel("Predicted Class")
plt.ylabel("Actual Class")

# Add numbers inside boxes
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

# Color scale
plt.colorbar()

plt.show()


print("\nPrecision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
